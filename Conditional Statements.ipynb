```python
total_units = int(input("Enter total units consumed:"))

if total_units <= 100:
    cost = total_units * 5 
elif total_units <= 300:
    cost = (100 * 5) + (total_units - 100) * 8
else:
    cost = (100 * 5) + (200 * 8) + (total_units - 300) * 12
print("Total Bill: Rs.", cost)
```
