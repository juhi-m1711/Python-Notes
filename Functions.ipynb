```python
def total(qty, price):
    t = qty * price
    return t
```

```python
total(12,12)
```

```python
def area_triangle(base, height):
    a = 0.5 * base * height
    return a
```

```python
area_triangle(2, 4)
```

```python
def is_even(num):
    e = num % 2 == 0
    return e
```

```python
is_even(3)
```

```python
def is_even_num(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
is_even_num(8)
```

# Lambda Function

```python
square = lambda s : s * s 
```

```python
square(4)
```

```python
rect = lambda l, b: l * b
```

```python
rect(6,3)
```

```python
iseven = lambda n : n % 2 == 0
```

```python
iseven(2)
```

```python
first_char = lambda name: name[0]
```

```python
first_char("Juhi")
```

```python
nums = [1,2,3,4]
squares = list(map(lambda x : x ** 2, nums))
```

```python
squares
```

```python
nums.extend([5,6,7,8,9])
nums
```

```python
greater_than_5 = list(filter(lambda x: x>5, nums))
```

```python
greater_than_5
```

```python
words = ["apple", "banana", "chikoo", "durian", "eggplant", "kiwi"]
sorted_words = sorted(words, key= lambda words: len(words))
```

```python
sorted_words
```

```python
sorted_num = sorted(nums, key= lambda nums: nums)
```

```python
sorted_num
```

```python
def square(n):
    return n * n 
result = square(4)

print(result + 10)
```

```python
def salary(base_salary):
    def bonus(bonus_percent):
        total_salary = base_salary + (base_salary * bonus_percent / 100)
        return total_salary
    return bonus
```

```python
e1 = salary(4000)
e1(10)
```
