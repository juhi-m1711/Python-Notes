# Insert values into strings

Using % 

```python
n = "Juhi"
a = 21
```

```python
print("My name is %s and I am %i years old." %(n,a)) # %s represents string and %i represents integer, write %(n,a) in order with the values.
```

Using FORMAT function

```python
print("My name is {} and I am {} years old.".format(n,a)) #format function 
```

```python
print("My name is {1} and I am {0} years old.".format(a,n)) # uses indexing if variables are not written in order
```

Using f string

```python
print(f"My name is {n} and I am {a} years old.") # use f before starting the string and add the variable names directly inside {}
```
