3. Set (Unordered, Unique Items Only)

```python
# Passing a set using {}
s = {"google", "amazon", "delloite", "kpmg"}
s
```

```python
s.add("ey") # adds a value 
```

```python
s
```

```python
s.remove("ey") # removes a value, can give an error if value not found
```

```python
s
```

```python
s.discard("kmpo") # remove a value, does not give any error
```

```python
s
```

```python
x = {1,2,3,4,5,6}
y = {2,3,4,5,6,7,8,9,0}
```

```python
x.union(y) # combines the set
```

```python
x.intersection(y) # results the common elements
```

```python
x.difference(y) # items in x not in y
```

```python
y.difference(x) # items in y not in x
```

```python
x.symmetric_difference(y) # different values in both set
```

```python
x.update([10,11,12]) # adds value to a set
```

```python
x
```

4. Dictionary (Key: Value Pairs, Ordered, Unique keys, Mutable)

```python
# Passing a dictionary using {}
data = {"names" : ["Juhi", "Nisha", "Aryan"], "age" : [ 21, 25, 23 ]}
```

```python
data
```

```python
data.get("names") # Safe key access
```

```python
data.keys() # Gets all the keys of a dict
```

```python
data.values() # Gets all the values of a dict
```

```python
data.items() # Gets all the items of a dict
```

```python
data.update({"grades" : ["A", "B", "A"]}) # Merge another dictionary
```

```python
data
```

```python
data.pop("age") # Remove key and return value
```

```python
data
```

```python
data.popitem() # Removes last inserted item
```

```python
data
```

```python
new_data = data.copy() # Makes copy
```

```python
new_data
```

```python
new_data.setdefault("city","Nagpur") # Get or set default
```

```python
d = new_data.fromkeys(["math", "science"], "pass") # Create new dict from list of keys
```

```python
d
```
