```python
x < y # compares if x is smaller than y
```

4. Logical Operators

```python
p = 102 
q = 51
```

```python
p > q and p == q # 'and' return TRUE if both conditions are true, else returns FALSE 
```

```python
p > q or p == q # 'or' returns TRUE if either of the conditions return true, if both conditions are false it returns FALSE
```

```python
not p > q # 'not' Reverses the result. 'not' can be used as a unary operator
```

5. Membership Operators

```python
fruit = "apple"
```

```python
'a' in fruit # True if value is present
```

```python
'a' not in fruit # True if value is not present
```
