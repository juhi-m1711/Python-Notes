# 1. While Loop

```python
a = 1
while a < 10:
    print (a)
    a +=1
```

```python
b = 1
while b < 10:
    b +=1
    print(b)
```

```python
c = 10
while c >= 1:
    print (c)
    c -=1
```

# 2. For Loop

```python
for i in range(5):
    print(i)
```

```python
for i in range(5, 15):
    print(i)
```

```python
for i in range(5, 15, 2):
    print(i)
```

```python
for i in range(15, 4, -1):
    print(i)
```

# Loop Control: continue, break, pass

```python
numbers = [1,2,-3,-4,-9,8,6]

for num in numbers:
    if num < 0:
        continue # Skips negative numbers
    print(num)
```

```python
for i in range(2,20):
    if i % 2 == 0: # Skips Even Numbers
        continue
    print(i)
```

```python
for i in range(2,20):
    if i % 2 != 0: # Skips Odd Numbers
        continue
    print(i)
```

```python
i = 0

while i < 12:
    print(i)
    if i == 5:
        break # Breaks the loop when i reaches five
    i +=1
```

```python
words = ['Is', 'Sun', 'Planet', 'Or', 'Star']

for w in words:
    if len(w) > 2:
        continue
    print (w)
```

# Zip Function

```python
# Zipping 
names = ["Amit", "Sneha", "Ravi"]
scores = [96,85,86]

for x, y in zip(names, scores):
    print(x, " scored", y, " marks.")
```

```python
# Unzipping
zipped_data = [("Amit", 96), ("Sneha", 85), ("Ravi", 86)]
x, y = zip(*zipped_data)
```
